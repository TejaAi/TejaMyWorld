```python
import evalml
import pandas as pd
```

```python
df = pd.read_csv("Avenger_Assemble.csv")
df
```

```python
df.drop(columns=["Unnamed: 0"],inplace=True)
```

```python
df
```

```python
#Removing the Unwanted Symbols & Changing the Datatypes
df["bike_price"]=df["bike_price"].str.replace("[\₹\,]","",regex=True)
df['km_driven']=df['km_driven'].str.replace("km","")
df['km_driven']=df['km_driven'].str.replace(",","")
df
```

```python
#Adding the Row ["City"]

new =df["bike_address"].str.split(",",n=3,expand=True)
df["city"]=new[1]
df
```

```python
df.drop(columns=["bike_address"],inplace=True)
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df = df.drop_duplicates()
df
```

```python
df.isnull().sum()
```

```python
df['km_driven']=df['km_driven'].str.replace(".","")
```

```python
df["km_driven"]=df["km_driven"].astype(int)
```

```python
df.dtypes
```

```python
df.iloc[:,1:-5]
```

```python
po = df.iloc[:,1:-5]
y=po.squeeze()
y
```

```python
X=df.drop(columns=["km_driven"])
```

```python
X
```

```python
X_train,X_test,y_train,y_test=evalml.preprocessing.split_data(X,y,problem_type='regression')
```

```python
import evalml
evalml.problem_types.ProblemTypes.all_problem_types
```

```python
from evalml.automl import  AutoMLSearch
automl=AutoMLSearch(X_train=X_train,y_train=y_train,problem_type='regression')
automl.search()
```

```python
automl.rankings
```

```python
automl.best_pipeline
```

```python
automl.describe_pipeline(automl.rankings.iloc[0]["id"])
```

```python
best_pipeline=automl.best_pipeline
```

```python
best_pipeline.score(X_test,y_test,objectives=['r2'])
```

```python
best_pipeline.save('olx.pkl')
```
