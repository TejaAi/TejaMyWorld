```python
import pandas as pd
olx =pd.read_csv('olxPunjab.csv')
```

```python
olx.drop(columns=["Unnamed: 0"],inplace=True)
```

```python
import selenium
from selenium import webdriver as wb
chr1 = wb.Chrome('C:\Program Files (x86)\Google\Chrome\Application\chromedriver.exe')
```

```python
from tqdm import tqdm
import numpy
alldetails=[]
erroitems=[]
for t in tqdm(olx.Links):
    chr1.get(t)
    try:
        try:
            Price=chr1.find_element_by_class_name('_2xKfz').text
        except:
            Price='NULL'
        try:
            Brand=chr1.find_element_by_class_name('_2vNpt').text
        except:
            Brand="NULL"
        try:
            Model=chr1.find_elements_by_class_name('_2vNpt')[-3].text
        except:
            Model="NULL"
        try:
            Kilometers=chr1.find_elements_by_class_name('_2vNpt')[-1].text
        except:
            Kilometers="NULL"
        try:
            Manfacture_year=chr1.find_elements_by_class_name('_2vNpt')[-2].text
        except:
            Manfacture_year="NULL"
        try:
            Join_of_candidate=chr1.find_element_by_class_name('_3AdUv')
            Join_of_candidate1=Join_of_candidate.find_elements_by_tag_name('span')
            Join_of_candidate=Join_of_candidate1[-1].text
        except:
            Join_of_candidate="NULL"
        try:
            Location=chr1.find_elements_by_class_name('_2FRXm')[-2].text
        except:
            Location='NULL'
        tempa ={'Bike_price':Price,
                'Km_driven':Kilometers,
                'Make_year':Manfacture_year,
                'Bike_name':Brand,
                'Bike_model':Model,
                "Bike_address":Location,
                'Seller_info':Join_of_candidate,
            }
        alldetails.append(tempa)
    except:
        erroitems.append(t)
        
    
```

```python
len(alldetails)
```

```python
dlf=pd.DataFrame((alldetails))
dlf.to_csv('Olx_Punjab.csv')
```
