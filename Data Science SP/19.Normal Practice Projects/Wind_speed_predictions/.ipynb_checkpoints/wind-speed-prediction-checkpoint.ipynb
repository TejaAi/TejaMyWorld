```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
df = pd.read_csv('/kaggle/input/wind-speed-prediction-dataset/wind_dataset.csv', sep = ',')
#df = pd.DataFrame(df)
df.head()
```

# Converting Date to date time format

```python
df['DATE'] = pd.to_datetime(df['DATE'])
df.DATE.dtype
```

# Feature Engineering on Date

```python
df['year'] = df['DATE'].dt.year
df['month'] = df['DATE'].dt.month
df['day'] = df['DATE'].dt.day
df['week'] = df['DATE'].dt.isocalendar().week
df['qtr'] = df['DATE'].dt.quarter
df['weekday'] = df['DATE'].dt.weekday
```

```python
df.head()
```

```python
df.weekday.nunique()
```

```python
df.isnull().sum()
```

```python
df.info()
```

# Get Dummies for week column

```python
df = pd.get_dummies(df, columns = ['week'])
df = df.drop(['DATE'], axis = 1)
df.head()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
sc = StandardScaler()
```

# Filling Null columns

```python
df.fillna(df.median()[0], inplace= True)
```

```python
df.isnull().sum()
```

```python
X = df.drop(['WIND'], axis = 1)
y = df['WIND']
X = sc.fit_transform(X)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 44)
```

```python
import lightgbm as lgb
hyper_params = {
    'task': 'train',
    'boosting_type': 'gbdt',
    'objective': 'regression',
    'metric': ['l1','l2'],
    'learning_rate': 0.005,
    'feature_fraction': 0.9,
    'bagging_fraction': 0.7,
    'bagging_freq': 10,
    'verbose': 0,
    "max_depth": 8,
    "num_leaves": 12,  
    "max_bin": 512
}
gbm = lgb.LGBMRegressor(**hyper_params)
from sklearn.metrics import mean_squared_log_error
from sklearn import metrics
from sklearn.metrics import r2_score, mean_squared_error
```

```python
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor, ExtraTreesRegressor, AdaBoostRegressor, BaggingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neural_network import MLPRegressor
mlp = MLPRegressor(early_stopping=True)
br = BaggingRegressor()
import catboost as cb
import shap
cat = cb.CatBoostRegressor(loss_function='RMSE', random_state = 50,  iterations=1000,learning_rate=0.01,depth=6,l2_leaf_reg=7)
from sklearn.experimental import enable_hist_gradient_boosting 
from sklearn.ensemble import HistGradientBoostingRegressor
hgb = HistGradientBoostingRegressor(learning_rate=0.1, max_iter=2500, random_state=44)
dt = DecisionTreeRegressor(max_depth=None, min_samples_split=2,random_state=43)
rf = RandomForestRegressor(n_estimators=700, criterion='mse', max_depth=7, random_state = 41)
gbr = GradientBoostingRegressor()
```

# Modeling

```python
models = [cat, hgb, dt,rf,gbr,gbm]
for model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    r2_score = metrics.r2_score(y_test, y_pred)
    RMSE = np.sqrt(metrics.mean_squared_error(y_test,y_pred))
    print(model, '\n', 'RMSE:', RMSE, '\n', "R2Score:", '\n', r2_score, '\n')
```
