```python
df.columns
```

```python
visualize=df[[ 'Cylinders', 'Transmission','Fuel Consumption(Comb (L/100 km))','CO2 Emissions(g/km)', 'CO2 Rating', 'Smog Rating']]
visualize.hist(color='blue',figsize=(10,10))
plt.tight_layout();
plt.show()
```

```python
from pandas.plotting import scatter_matrix
plt.rc('font',size=12)
scatter_matrix(df[['Cylinders','Transmission','Fuel Consumption(Comb (L/100 km))','CO2 Emissions(g/km)', 'CO2 Rating', 'Smog Rating']], figsize=(20,20))
plt.tight_layout;
plt.show()
```

```python

g=sns.PairGrid(df[['Cylinders','Transmission','Fuel Consumption(Comb (L/100 km))','CO2 Emissions(g/km)', 'CO2 Rating', 'Smog Rating']], height=4, aspect=0.9)
g.map_upper(sns.regplot,scatter_kws={'color':'red'},line_kws={"color":"blue"})
g.map_lower(sns.kdeplot,cmap="plasma",shade=True,threshold=0.05)
g.map_diag(plt.hist,color="Yellow")
plt.tight_layout;
plt.show();
```

```python
corr=df.drop('Model Year',axis=1).corr()
corr
```

```python
corr_top=corr[abs(corr)>=.8]
plt.figure(figsize=(12,12))
sns.heatmap(corr_top,cmap="cividis_r",annot = True)
```
