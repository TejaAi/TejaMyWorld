```python
df.rename(columns={'Fuel Consumption(Comb (L/100 km))':"Fuel_consuption",'CO2 Emissions(g/km)':'co2','Engine Size(L)':'ES'},inplace=True)
```

```python
df.columns
```

```python
df.drop(columns=['Fuel Consumption(Hwy (L/100 km))','Fuel Consumption(Comb (mpg))','Fuel Consumption (City (L/100 km)'],inplace=True)
```

```python
df.columns
```

```python
from feature_engine.outliers import Winsorizer
```

```python
win_E = Winsorizer(capping_method="iqr",tail='both',fold=1.5,variables=["ES"])
```

```python
win_C = Winsorizer(capping_method="iqr",tail='both',fold=1.5,variables=["Cylinders"])
```

```python
win_CO = Winsorizer(capping_method="iqr",tail='both',fold=1.5,variables=["co2"])
```

```python
win_CO2 = Winsorizer(capping_method="iqr",tail='both',fold=1.5,variables=["CO2 Rating"])
```

```python
df["ES"] = win_E.fit_transform(df[["ES"]])
```

```python
df["Cylinders"] = win_C.fit_transform(df[["Cylinders"]])
```

```python
df["co2"]=win_CO.fit_transform(df[["co2"]])
```

```python
df["CO2 Rating"] =win_CO2.fit_transform(df[["CO2 Rating"]])
```

### Encodeing

```python
from sklearn.preprocessing import LabelEncoder
```

```python
from sklearn.preprocessing import OneHotEncoder
```
